```python
import networkx as nx
import pickle
```

```python
with open('../data/similarity_network.pickle', 'rb') as f:
     sim_G = pickle.load(f)[0]
```

```python
cluster_dict = {}

for node in sim_G.nodes():
    cluster = sim_G.nodes[node]['cluster']
    if cluster in cluster_dict:
        cluster_dict[cluster].add(node)
    else:
        cluster_dict[cluster] = {node}
```

```python
cluster_dict
```

```python
with open('../data/retweet_geonetwork_russian.pickle', 'rb') as f:
     retweet_G = pickle.load(f)
```

```python
for node in retweet_G.nodes():
    for cluster, links in cluster_dict.items():
        if retweet_G.nodes[node]['url'] in links:
            retweet_G.nodes[node]['cluster'] = f"Cluster {cluster}"
```

```python
with open('../new_graphs/retweet_geonetwork_russian.pickle', 'wb') as f:
     nx.write_gpickle(retweet_G, f, protocol=pickle.HIGHEST_PROTOCOL)
```
